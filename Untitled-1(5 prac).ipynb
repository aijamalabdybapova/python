Метод describe возвращает основные статические показатели для каждого чисоового столбца, такие как количество, среднее значение, стандартное отклонение, минимальное и максимальное значение, медиана и квартили.
В выводе метода describe() в Pandas представлены следующие статистические показатели для каждого числового столбца:
count: В кажлом столбце по 887 непустых значений.
mean: Среднее арифметическое значение. Например средее значение возраста пассажиров равен 29.471443, а средним  значением пассажирского класса - 2.305524.
std: Стандартное отклонение. Например, Для столбца "Age" значение стандартного отклонения равно 14.121908. Это указывает на разброс возраста пассажиров относительно их среднего возраста.
Для столбца "Siblings/Spouses Aboard" значение стандартного отклонения равно 1.104669. Это означает, что количество братьев/сестер или супругов, путешествующих с пассажиром, отклоняется от среднего значения на примерно 1.104669.
min: Минимальное значение в столбце. Например, самым молодым пассажиром является младенец с возрастом 0.42, а минимальное значение в стобце Pclass 1(высший класс.)
25% (первый квартиль): Значение, которое разделяет самые низкие 25% данных. Это также известно как нижний квартиль.
50% (медиана): Значение, которое разделяет данные пополам. То есть, 50% значений ниже медианы, и 50% значений выше.
75% (третий квартиль): Значение, которое разделяет самые низкие 75% данных. Это также известно как верхний квартиль.
max: Максимальное значение в столбце. Например, максимальным количеством братьев, сестер или супругов, находящихся на борту с пассажиром - 8 человек, а количество родителей или детей, находящихся на борту с пассажиром - 6 человек
