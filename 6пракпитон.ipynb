Цель: на данной практической работе необходимо использовать датасет с прошлой практической работы и визуализировать данные в виде графиков

Мы импортировали с помощью функции read_csv и получился датасет по наименованию переменной

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas  as pd
titanic= pd.read_csv(r"C:\Users\User\Desktop\titanic.csv")
titanic
```

Создаем новую фигуру (график) с заданным размером. Строим график линии, где по оси x отображаются последние 5 значений столбца 'Name', а по оси y - последние 5 значений столбца 'Age'. label используется для создания метки легенды. Устанавливаем заголовок для графика.Устанавливаем подписи осей x и y соответственно. Добавляем легенду, которая в данном случае будет отображать метку 'Возраст пассажира'.
Таким образом, этот код создает линейный график, отражающий статистику возраста 5 пассажиров в датасете. Легенда указывает, что график представляет возраст пассажира.

```python
plt.figure(figsize=(20,10))
plt.plot(titanic['Name'].tail(5),titanic['Age'].tail(5), label = 'Возраст пассажиров')
plt.title('Статистика возвраста пассажиров')
plt.xlabel('Имя пассажира')
plt.ylabel('Возраст')
plt.legend()
plt.show()
```

Создаем новую фигуру (график) с заданным размером. Строим линейный график, где по оси x отображаются последние 10 значений столбца 'Name', а по оси y - последние 10 значений столбца 'Siblings/Spouses Aboard'. Опция label создает метку для легенды, color="red" устанавливает цвет графика, marker='o' добавляет круглые маркеры, и markersize=7 устанавливает размер маркеров.Устанавливаем заголовок для графика. Устанавливаем подписи осей x и y соответственно.Добавляем легенду, которая в данном случае будет отображать метку 'Количество братьев и сестер/супруги на борту'.
Добавляем сетку на график для улучшения читаемости.
Таким образом, этот код создает линейный график, отражающий статистику количества братьев и сестер/супруги на борту последних 10 пассажиров в датасете. Маркеры 'o' позволяют визуально выделить отдельные точки, и сетка делает график более наглядным.

```python
plt.figure(figsize=(20,10))
plt.plot(titanic['Name'].tail(10),titanic['Siblings/Spouses Aboard'].tail(10), label = 'Количество братьев и сестер/супруги на борту', color = "red", marker='o', markersize=7)
plt.title('Статистика количества братьев и сестер/супруги на борту')
plt.xlabel('Имя пассажира')
plt.ylabel('Количество братьев и сестер/супруги на борту')
plt.legend()
plt.grid()
plt.show()
```

Создаем новую фигуру (график) с заданным размером. Строим линейный график, где по оси x отображаются последние 8 значений столбца 'Name', а по оси y - последние 8 значений столбца 'Age'. Добавляем столбчатую диаграмму поверх линейного графика. Опция label создает метку для легенды, а color='tan' устанавливает цвет столбцов.Устанавливаем заголовок для графика. Устанавливаем подписи осей x и y соответственно.
Добавляем легенду, которая будет содержать две метки: 'Возраст пассажира' для линейного графика и 'Возраст пассажира столбцы' для столбчатой диаграммы.
Таким образом, мы получаем гибридный график, который позволяет визуализировать возраст пассажиров как с использованием линий, так и столбцов.

```python
plt.figure(figsize=(20,10))
plt.plot(titanic['Name'].tail(8),titanic['Age'].tail(8), label = 'Возраст пассажира', color = "black")
plt.bar(titanic['Name'].tail(8),titanic['Age'].tail(8), label = 'Возраст пассажира столбцы', color = 'tan')
plt.title('Статистика возвраста пассажиров')
plt.xlabel('Имя пассажира')
plt.ylabel('Возраст')
plt.legend()
plt.show()
```

Код создает столбчатую диаграмму для отображения  последних 6 пассажиров из датасета. Ось x представляет имена пассажиров, ось y - 'Родители/дети на борту', . Цвет столбцов - розовый.

```python
plt.figure(figsize=(20,10))
plt.bar(titanic['Name'].tail(6),titanic['Parents/Children Aboard'].tail(6), label = 'Родители/дети на борту', color = 'pink')
plt.title('Статистика количества родителей/детей на борту, путешествующих с пассажиром' )
plt.xlabel('Имя пассажира')
plt.ylabel('Родители/дети на борту')
plt.legend()
plt.show()
```

Код визуализирует статистику пассажирских классов у последних 8 пассажиров с использованием линейного графика и столбчатой диаграммы для отображения информации об количестве братьев и сестер/супруг на борту. Также есть настройки для размера фигуры, цвета, маркеров и других параметров

```python
plt.figure(figsize=(20,10))
plt.plot(titanic['Name'].tail(8),titanic['Pclass'].tail(8), label = 'Пассажирский класс', color = "orange", markersize=7, marker='o')
plt.bar(titanic['Name'].tail(8),titanic['Siblings/Spouses Aboard'].tail(8), label = 'Количество братьев и сестер/супруги на борту, путешествующих с пассажирм', color = 'coral')
plt.title('Статистика пассажирских классов')
plt.xlabel('Имя пассажира')
plt.ylabel('Количество братьев и сестер/супруги на борту')
plt.legend()
plt.show()
```

Данный код визуализирует сравнение возраста и пассажирского класса последних 7 пассажиров на корабле "Титаник". График представлен в виде столбчатой диаграммы, где по оси X отображаются имена пассажиров, а по оси Y - значения возраста и пассажирского класса. Каждый пассажир представлен двумя столбцами: один столбец отображает возраст, а другой - пассажирский класс. Цвета столбцов различны: возраст показан розовым цветом, а пассажирский класс - фиолетовым.

```python
plt.figure(figsize=(20,10))
plt.bar(titanic['Name'].tail(7),titanic['Age'].tail(7), label = 'Возраст пассажира', color = 'pink')
plt.bar(titanic['Name'].tail(7),titanic['Pclass'].tail(7), label = 'Пассажирский класс', color = 'purple')
plt.legend()
plt.xlabel('Имя пассажира')
plt.ylabel('Пассажирский класс')
plt.show()
```

Данный код визуализирует соотношение выживших и погибших пассажиров на корабле Титаник. Он использует круговую диаграмму, где каждый сектор представляет одну из категорий выживания (0 - погибшие, 1 - выжившие). Процентное значение отображается внутри каждого сектора. В данном случае, оранжевый сектор соответствует погибшим пассажирам, а розовый - выжившим.

```python
Survived  = titanic['Survived'].value_counts()
plt.pie(Survived, labels=Survived.index, autopct='%1.1f%%', colors = ['orange','pink'])
plt.title('Соотношение выживания')
plt.show()
```

Данный код визуализирует соотношение возрастов пассажиров на титанике в диапазоне от 18 до 23 лет. Круговая диаграмма показывает процентное соотношение каждого возраста от общего числа пассажиров в данном диапазоне. Каждому возрасту соответствует свой цвет на диаграмме.

```python
Age = titanic['Age'].value_counts()
valid_Age = Age[(Age.index >= 18) & (Age.index <= 23)]
labels = valid_Age.index
plt.pie(valid_Age, labels=labels, autopct='%1.1f%%', colors = ['pink','tan','lightyellow','lightgreen','lightblue','purple','coral'])
plt.title('Возрасты')
plt.show()
```

Данный код визуализирует связь между возрастом пассажиров и их именами на графике типа "scatter plot". Каждая точка на графике представляет собой пассажира, где по оси X отображается их возраст, а по оси Y - их имя. Цвет точек задан как фиолетовый. График позволяет наглядно увидеть распределение возрастов пассажиров и их соответствующие имена.

```python
plt.figure(figsize=(10,9))
plt.scatter(titanic['Age'].head(40), titanic['Name'].head(40), color = 'purple')
plt.title('Связь между пассажиром и возрастом')
plt.xlabel('Возраст')
plt.ylabel('пассажир')
plt.show()
```

Данный код визуализирует связь между пассажиром и стоимостью билета на примере первых 50 пассажиров. Ось x представляет собой стоимость билета, а ось y - имена пассажиров. Каждая точка на графике представляет одного пассажира и его стоимость билета. Цвет точек задан как красный. Таким образом, график позволяет наглядно увидеть распределение стоимости билетов среди первых 50 пассажиров и их соответствующие имена.

```python
plt.figure(figsize=(10,10))
plt.scatter(titanic['Fare'].head(50), titanic['Name'].head(50), color = 'red')
plt.title('Связь между пассажиром и стоимостью билета')
plt.xlabel('Стоимсоть билета')
plt.ylabel('Пассажир')
plt.show()
```

Данный код визуализирует гистограмму распределения возраста пассажиров на корабле Титаник. Ось X представляет собой диапазон возрастов, разделенный на 30 интервалов (bins), а ось Y показывает количество пассажиров, попадающих в каждый интервал. Цвет гистограммы - оранжевый.

```python
plt.figure(figsize=(18,10))
plt.hist(titanic['Age'], bins=30, color = ["orange"], edgecolor = 'coral', alpha=0.7, label = 'Возраст')
plt.legend()
plt.show()
```

Данный код визуализирует гистограмму распределения пассажирского класса на титанике. На оси X отображаются значения пассажирского класса, а на оси Y - количество пассажиров в каждом классе. Цвет столбцов гистограммы - светло-зеленый.

```python
plt.figure(figsize=(18,10))
plt.hist(titanic['Pclass'], bins=10, color = ["lightgreen"], edgecolor = 'black', alpha=0.7, label = 'Пассажирский класс')
plt.legend()
plt.show()
```

Данный код визуализирует гистограмму, которая показывает распределение количества родителей/детей на борту, путешествующих с пассажиром на корабле Титаник. Ось x представляет собой количество родителей/детей, а ось y - количество пассажиров с определенным количеством родителей/детей.

```python
plt.figure(figsize=(18,10))
plt.hist(titanic['Parents/Children Aboard'], bins=30, color = ["lightblue"], edgecolor = 'blue', alpha=0.7, label = 'Количество родителей/детей  на борту, путешествующих с пассажиром')
plt.legend()
plt.show()
```
